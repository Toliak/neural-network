```python
import Dataset
import numpy as np
from Utils import visualize
from Utils import image_to_grayscale

dataset_list = Dataset.get_full_dataset()
INDEX = 0
data = dataset_list[INDEX]
image, title = data['image'], data['title']

image = image_to_grayscale(image)
    
visualize(image, f'grayscale {title}')
```

```python
from skimage.filters import threshold_isodata, threshold_otsu, threshold_li, threshold_mean, \
    threshold_minimum, threshold_triangle, threshold_yen, threshold_niblack
from scipy.ndimage import label

threshold_ways = (
    # threshold_isodata,
    # threshold_li,
    threshold_mean,
    # threshold_minimum,
    # threshold_otsu,
    # threshold_triangle,
    # threshold_yen,
    # threshold_niblack,
)

for filter_function in threshold_ways:
    threshold_value = filter_function(image)
    binary = image > threshold_value
    
    labeled, amount = label(binary)
    visualize(labeled, f'labeled {title} {filter_function.__name__}')
    print(f'amount {title} {filter_function.__name__}: {amount}')
```
