```python
from sklearn.mixture import GaussianMixture
import Dataset
from Utils import visualize
import matplotlib.pyplot as plot
import numpy as np

original = Dataset.load_by_path('small.png')
visualize(original, 'original')
```

```python
newdata = original.reshape(original.shape[0] * original.shape[1], original.shape[2])
gm_class = GaussianMixture(n_components=3, covariance_type="tied")
gmm = gm_class.fit(newdata)

cluster = gmm.predict(newdata)
cluster = cluster.reshape(*original.shape[:-1])
visualize(cluster, 'cluster')
```

```python
crop = original.copy()
crop[cluster != 1] = 0

visualize(crop, 'crop')
```

```python
newdata = crop.reshape(crop.shape[0] * crop.shape[1], crop.shape[2])
gm_class = GaussianMixture(n_components=3, covariance_type="tied")
gmm = gm_class.fit(newdata)

cluster = gmm.predict(newdata)
cluster = cluster.reshape(*crop.shape[:-1])
visualize(cluster, 'cluster')
```
